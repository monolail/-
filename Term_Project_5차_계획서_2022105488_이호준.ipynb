```python
#4 홈팀과 어웨이팀의 승패 (*위의 팀 알림이에서 복사해서 붙여 넣어주세요*)
print("보고싶은 팀을 선택해 주세요!!")
home_Team = input("Home Team : ")
away_Team = input("Away Team : ")
print("Selected team(Home : Away) : " + home_Team + " : " + away_Team)

result1 = pd.concat([data[data['HomeTeam']== home_Team], data[data['AwayTeam']== home_Team]]).sort_index()
result2 = pd.concat([result1[result1['AwayTeam']==away_Team]]).sort_index()
result2
```

```python
#5 data(PL_result) 정보 확인
data.info()
data.isna().sum()
```

```python
#6 불필요한 열의 내용은 제거한다.
data.drop(['DateTime','HTHG','HTAG','HTR','Referee', 'HS', 'AS','HST','AST', 'HC', 'AC', 'HF','AF', 'HY', 'AY', 'HR', 'AR']
              , axis=1, inplace=True)
data
```

```python
#7 새로운 열 내용으로 채우기
data.replace({'H' : 0, 'D' : 1, 'A' : 2}, inplace=True)
data
```

```python
#8 unique를 통해 개별 컬럼 내에 유일한 값을 리스트화
team_name = list(data["HomeTeam"].unique())

data_team = pd.DataFrame(columns=team_name, index=["G", "D"])

#8 골득실 data_team에 널기
for i in team_name:
    data_team[i].G = sum(data["FTHG"][data["HomeTeam"]== i])
    data_team[i].G += sum(data["FTAG"][data["AwayTeam"] == i])
    data_team[i].D = sum(data["FTAG"][data["HomeTeam"] == i])
    data_team[i].G += sum(data["FTHG"][data["AwayTeam"] == i])
#8 팀순서는 알파베 ABC 순서   
data_team
```

```python
#8 전치 행렬 전환한다.
data_team = data_team.T
data_team
```

```python
data_team = data_team.reset_index()
data_team.rename(columns = {'index' : 'team'}, inplace = True)
df_team_G = data_team.sort_values("G", ascending=False)
```

```python
#9 seaborn을 이용해서 골득실 그래프로 만들기(한글 깨짐 현상으로 인한 영어 사용)
figure, (ax1,ax2) = plt.subplots(nrows = 1, ncols =2 ) 
figure.set_size_inches(18,12)
sns.barplot(data=df_team_G, x="G", y="team", ax=ax1)
sns.barplot(data=df_team_G, x="D", y="team", ax=ax2)
ax1.set_xlabel("Goal", fontsize=30)
ax2.set_xlabel("Difference", fontsize=30)
```

```python
#10 승률 그래프제작 
cnt = 0

#10 새로운 열인 win rate 생성하고 삽입하기
data_team.insert(3, "Winrate", 0)

#10 2000년대 이후로 설정
i = 0
for i in range(0, 10802):
    season = (data["Season"].loc[i]).split("-")
    mm = int(season[0])
    if mm < 2000:
        data.drop(i, inplace=True)
#10 team이름의 홈팀에서 이긴 경기를 총 경기수로 나누고 100을 곱한 후 2자리에서 반올림한다.
for name in team_name:
    All_match = sum(data["HomeTeam"]==name) + sum(data["AwayTeam"]==name)
    Win_match = sum((data["HomeTeam"]==name) & (data["FTR"]==0))
    Win_match += sum((data["AwayTeam"]==name) & (data["FTR"]==2))
    if All_match == 0:
        continue
    data_team["Winrate"].loc[cnt] = round((Win_match/All_match) * 100, 1)
    cnt+=1
    
data_team
```

```python
#11 seaborn을 통해 승률 그래프 제작

figure, (ax1,ax2) = plt.subplots(nrows = 1, ncols =2 ) 
figure.set_size_inches(18,12)
sns.barplot(data=df_team_G, x="G", y="team", ax=ax1)
sns.barplot(data=data_team.sort_values("Winrate", ascending=False), x="Winrate", y="team", ax=ax2)
ax1.set_xlabel("Goal", fontsize=30)
ax2.set_xlabel("Win rate", fontsize=30)
```
